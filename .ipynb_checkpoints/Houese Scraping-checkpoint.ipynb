```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
Propertys = []
Property_data=[]
for page in range(581):
    response = requests.get(f'https://realestate.eg/en/for-sale?page={page}')
    soup = BeautifulSoup(response.text, "html.parser")
    print(response)
    for Property in soup.find_all('div',class_='property-card'):
        try:
            Property_price_EGP = float(Property.find('span',class_='number').text.replace(' EGP','').replace(',',''))
        except:
            Property_price_EGP = ''
        try:
            Location_City = Property.find('div',class_='location').select('span')[0].text
        except:
            Location_City = ''
        try:
            Property_Type = Property.find('div',class_='info').select('li')[0].text 
        except:
            Property_Type = ''
        try:
            Bed_Rooms = int(Property.find('div',class_='info').select('li')[1].text)
        except:
            Bed_Rooms = ''
        try:
            Bath_Rooms = int(Property.find('div',class_='info').select('li')[2].text)
        except:
            Bath_Rooms = ''
        try:
            Home_Space_SQM = int(Property.find('div',class_='info').select('li')[3].text.replace('m2','').strip())
        except:
            Home_Space_SQM = ''
        
        Property_data = [Property_price_EGP,Location_City,Property_Type,Bed_Rooms,Bath_Rooms,Home_Space_SQM]
        Propertys.append(Property_data)
print('The Scraping Is Done')
```

```python
len(Propertys)
```

```python
Details = ['Property_price_EGP','Location_City','Property_Type','Bed_Rooms','Bath_Rooms','Home_Space_SQM']  
with open('houses_Egypt(1).csv', 'w', encoding='utf-8', newline ='') as file:
    write = csv.writer(file) 
    write.writerow(Details) 
    write.writerows(Propertys)
```

```python
Propertys = []
Property_data=[]
for page in range(37):
    response = requests.get(f'https://www.isqan.com/en/properties/buy/properties-for-sale-in-egypt?page={page}&pageSize=100')
    soup = BeautifulSoup(response.text, "html.parser")
    print(response)
    for Property in soup.find_all('div',class_='property-info'):
        try:
            Property_price_EGP = float(Property.find('div',class_='row align-items-center mb-2').select('p')[0].text.replace('EGP ','').replace(',','').strip())
        except:
            Property_price_EGP = ''
        try:
            Location_City = Property.find('p',class_='h5 text-truncate').text.split(',')[-2].strip()
        except:
            Location_City = ''
        try:
            Property_Type = Property.find('div',class_='row align-items-center').select('p')[1].text.split(',')[0].strip() 
        except:
            Property_Type = ''
        try:
            Bed_Rooms = int(Property.find('div',class_='row property-info__rooms pl-1 ng-star-inserted').select('p')[0].text.split(',')[0].strip())
        except:
            Bed_Rooms = ''
        try:
            Bath_Rooms = int(Property.find('div',class_='row property-info__rooms pl-1 ng-star-inserted').select('p')[1].text.split(',')[0].strip())
        except:
            Bath_Rooms = ''
        try:
            Home_Space_SQM = int(Property.find('div',class_='row property-info__rooms pl-1 ng-star-inserted').select('p')[2].text.split(',')[0].split(' ')[1])
        except:
            Home_Space_SQM = ''
        
        Property_data = [Property_price_EGP,Location_City,Property_Type,Bed_Rooms,Bath_Rooms,Home_Space_SQM]
        Propertys.append(Property_data)
print('The Scraping Is Done')
```

```python
Details = ['Property_price_EGP','Location_City','Property_Type','Bed_Rooms','Bath_Rooms','Home_Space_SQM']  
with open('houses_Egypt(2).csv', 'w', encoding='utf-8', newline ='') as file:
    write = csv.writer(file) 
    write.writerow(Details) 
    write.writerows(Propertys)
```

```python
Apartments_Properties = []
Apartment_data=[]
for page in range(10):
    response = requests.get(f'https://www.olx.com.eg/en/properties/commercial-for-sale/?page={page}&filter=delivery_date_eq_1%2Cfurnished_eq_2%2Cpayment_option_eq_1%2Ctype_eq_6')
    soup = BeautifulSoup(response.text, "html.parser")
    print(response)
    for Property in soup.find_all('li',class_='c46f3bfe'):
        try:
            Property_price_EGP = Property.find('div',class_='_52497c97').text.replace('EGP','').strip()
        except:
            Property_price_EGP = ''
        try:
            Location_governorate = Property.find('div',class_='a52608cc').select('span')[8].text.split(',')[1].replace('•','').strip()
        except:
            Location_governorate = ''
        try:
            Location_City = Property.find('div',class_='a52608cc').select('span')[8].text.split(',')[0].replace('•','').strip()
        except:
            Location_City = ''
        try:
            Bed_Rooms = Property.find('div',class_='a52608cc').select('span')[2].text
        except:
            Bed_Rooms = ''
        try:
            Bath_Rooms = Property.find('div',class_='a52608cc').select('span')[4].text
        except:
            Bath_Rooms = ''
        try:
            Home_Space_SQM = Property.find('div',class_='a52608cc').select('span')[7].text.replace('SQM','').strip()
        except:
            Home_Space_SQM = ''
        Furnished  = 'No'
        Property_Type = 'Apartment'
        Payment_Option = 'Cash'
        Delivery_Date  = 'Ready To Move'
       
        
        Apartment_data = [Property_price_EGP,Location_governorate,Location_City,Bed_Rooms,Bath_Rooms,Home_Space_SQM,Furnished,Property_Type,Payment_Option,Delivery_Date]
        Apartments_Properties.append(Apartment_data)
print('The Scraping Is Done')
```

```python
Details = ['Property_price_EGP','Location_governorate','Location_City','Bed_Rooms','Bath_Rooms','Home_Space_SQM','Furnished','Property_Type','Payment_Option','Delivery_Date']  
with open('houses_Price_Egypt(12).csv', 'w', encoding='utf-8', newline ='') as file:
    write = csv.writer(file) 
    write.writerow(Details) 
    write.writerows(Apartments_Properties)
```
